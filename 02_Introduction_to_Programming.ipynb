# How to create algorithms?

Think about the calculation of the __factorial__ as a simple problem which can be solved with the help of an algorithm. It might seem too easy to give this task to a computer but it is the right exercise to start with for a beginner. The creation of an algorithm is divided into several interdependent steps:

### Understanding the problem
Although it seems to be obvious at first glance, it is crucial to fully comprehend the problem. An good example thereof is the [__Solow Growth Model__](http://www.pitt.edu/~mgahagan/Solow.htm "Solow Growth Model"). Without an understanding of the Solow Growth Model, even an excellent programmer will have troubles in creating an algorithm to solve it. Therefore, it is necessary to understand all the underlying aspects of a task before solving it.<br>

<div class="alert alert-block alert-info">
Hence, before tackling the factorial problem, it is necessary to understand its fundamental concept in more detail. For instance, the factorial of 5 looks as follows:
<br><br>
$$5! = 5 \times 4 \times 3 \times 2 \times 1= 120$$
<br>
Mathematically, the factorial of a number is thus the product of all positive integers less than or equal to this number. Moreover, it needs to be considered that only positive integers are allowed and that the factorial of 0 is equal to 1.
</div><br>

One way of understanding the problem is to see the algorithm as a __production function__ which takes some information from the outside world as input and creates in turn a certain output. It is even possible to incorporate one production function into another to create a bigger program. In computer science, this structure is known as the [__Divide and Conquer Principle__](https://en.wikibooks.org/wiki/Algorithms/Divide_and_Conquer "Divide and Conquer Principle"). Imagine a very complex issue. It is easier to tackle this task by breaking it down into smaller problems which can be solved independently.<br>

<div class="alert alert-block alert-info">
Let us turn again to our factorial problem. The input from the outside world needs to be a positive integer. The output is the product of all positive integers less than or equal to this number. Furthermore, the calculation of the factorial can be divided into several sequent steps:
<ol>
<li>A positive integer as an input by the user</li>
<li>The definition of the integers which are multiplied</li>
<li>The multiplication of all the integers</li>
<li>The Displaying of the result</li>
</ol>
</div>

### Developing alternative solutions of the problem
A problem can be solved in different ways. Often every approach provides some advantages. Therefore, a comparison of the different methods allows to find the most efficient and most convenient one. On top, more options allow also more flexibility which might be helpful later on during the programming process as soon as the programmer recognises that his first approach is not appropriate. Moreover, the development of the alternative solution is a good exercise to understand the task in more detail.<br>

<div class="alert alert-block alert-info">
While the file [01_Factorial.ipynb](https://www.dropbox.com/s/eg6eq3y4ja0ixup/01_Factorial.ipynb?raw=1 "1st Exercise") provides one approach to solve the factorial problem; there are numerous other ways to tackle this task. One suggestion can be found in the code below. Even though you do not understand all of the code, it is important to recognize that even such a simple problem can be solved by various approaches which all lead to the same result. You can try it out and run the code.
</div>

```python
def factorial(n):
    num = 1
    while n >= 1:
        num = num * n
        n = n - 1
    return num
n = int(input('Number to compute factorial: '))
print('Factorial of {} is {}'.format(n,factorial(n)))
```

### Writing down the solution
It turns out to be very helpful to write down the problem in a way which can be understood easily and is not in a machine language. This might be in the form of diagrams, flowcharts or **pseudocode**. A pseudocode looks similar to a computer code but is not written in a text or code editor. A pseudocode corresponds to a description of the code which should be implemented.<br>
<div class="alert alert-block alert-info">
__Pseudocode__<br>
In the case of the factorial calculation, we can use the following pseudocode:
<pre><code><div style="background-color:rgb(217, 237, 247)">1: Read n from the keyboard 
2: f = 1
3: for i = 2,...,n do
4: f = f*i
5: end for
6: return Factorial of n is f
</div></code></pre>

**!¡!¡ Markdown language offer a neater way to write pseudocode.**

This pseudocode can be interpreted in the following way:<br>
__Line 1:__<br>
The first line asks the user for which number the factorial should be calculated. The user’s input is stored in the [__variable__](https://en.wikibooks.org/wiki/Python_Programming/Variables_and_Strings "Variable") n. A variable is an expression which stores values or data.<br>
__Line 2:__<br>
The second line defines the variable f and assigns it the value 1. Basically, we tell the computer that there is an expression which is called f that is now equal to 1.<br>
__Line 3 to 5:__<br>
These lines do a so-called [__for loop__](https://wiki.python.org/moin/ForLoop "for loop"). A for loop repeats an operation for a given number of times. In this case, it repeats the process for every integer from 2 to n. The second variable i takes another value with every iteration. Furthermore, line 4 defines that the variable f takes a new value with every iteration which is equal to the old value of f times the i of that iteration.
<br>
__Line 6:__<br>After the last loop, the value of the variable f equals the factorial of n. The remaining line returns the value of f and thus shows the user the factorial of n.<br>
<br><br>

### Picking a programming language and coding the program
Depending on the problem, different programming languages are appropriate. However, in the extent of this lecture, we will stick to __Python__.
<div class="alert alert-block alert-info">
A programmed version of the factorial problem can be found in the file [01_Factorial.ipynb](https://www.dropbox.com/s/eg6eq3y4ja0ixup/01_Factorial.ipynb?raw=1 "1st Exercise"). As already mentioned, this code represents only one possible approach to a solution. Moreover, it is important to emphasize that this code requires an integer as input from the user. Otherwise, the algorithm will not work. You can try it out and use integers and strings (words) as input.</div>

```python
n = int(input('Number to compute factorial: '))
f = 1

for i in range(2, n+1):
    f = f * i

print('Factorial of {} is {}'.format(n,f))
```

There are three main ways to evaluate an algorithm; correctness, design and style. 

- Correctness: The program does what we want it to do. An example is if you create a code for computing a number, we evaluate how precise the program is. There is a distinction between partial correctness and total correctness. Partial correctness only requires the returned answer to be correct, and total correctness requires that the algorithm terminates.  

- Design: There are different opinions for how an algorithm should be designed. An algorithm can be correct, but poorly designed. Generally, an algorithm is designed well if the solution is done in a smart way, using resources in your computer effectively. For our purposes, the design does not have to be as perfect as a computer scientist, but it should still be fairly good. 

- Style: This is very important when the algorithm is read by 3rd persons. By style, we mean how easy it is to read the algorithm. It is important to use comments, and the layout of the code should be organized in a good way.
